```python
# Import Faker library for creating fake data
from faker import Faker
from datetime import datetime,timedelta
import random
```

```python
# Creating customer dataset
from faker import Faker
import pandas as pd
fake = Faker()
cities=['Delhi','Bhubaneswar','Mumbai','Pune',"Beijing",'srinagar','Kolkata','Hyderabad','Banglore','Chennai','Telengana','Kerala']


num_customers = 1000


customer_data = {
    'id': [i+1 for i in range(num_customers)],
    'name': [fake.name() for _ in range(num_customers)],
    'date_of_birth': [fake.date_of_birth(minimum_age=18, maximum_age=90).strftime('%d-%m-%Y') for _ in range(num_customers)],
    'email': [fake.email() for _ in range(num_customers)],
    'date_entered':[fake.date_time_between(start_date="-2y", end_date='now', ).strftime('%d-%m-%Y') for _ in range(num_customers)],
    'city': [random.choice(cities) for _ in range(num_customers)]
    
}


customer_df = pd.DataFrame(customer_data)

```

```python
customer_df
```

```python
#Covert to csv file
customer_df.to_csv('Raksha_customer.csv',index=False)
```

```python
# Creating Order dataset
fake = Faker()
num_orders = 1000
payment=['Credit card','Debit card','COD','Net Banking']
order_dates = [fake.date_time_between(start_date="-1y", end_date='now', ).strftime('%d-%m-%Y') for _ in range(num_orders)]
ship_dates = [(datetime.strptime(order_date, '%d-%m-%Y') + timedelta(days=random.randint(1, 5))).strftime('%d-%m-%Y') for order_date in order_dates]
delivery_dates = [(datetime.strptime(order_date, '%d-%m-%Y') + timedelta(days=random.randint(5, 15))).strftime('%d-%m-%Y') for order_date in order_dates]
order_data = {
    'orderid': [i for i in range(5000,6000)],
    'Orderdate': order_dates,
    'shipdate': ship_dates,
    'deliverydate': delivery_dates,
    'payment':[random.choice(payment) for i in range (num_orders)],
    'customer_id': [i+1 for i in range(num_orders)],
    'shipper_id': [np.random.randint(1, 6) for _ in range(num_orders)]
    
}

Order_df = pd.DataFrame(order_data)
```

```python
Order_df
```

```python
#Converting to csv file
Order_df.to_csv('Raksha_orders.csv',index=False)
```
